```python
import pandas as pd

excel_file = pd.ExcelFile("data.xlsx")

dfs = []
for sheet in excel_file.sheet_names:
    df = pd.read_excel(excel_file,sheet_name=sheet,header=1)
    df["연도"] = sheet
    dfs.append(df)

len(dfs)
```

```python
combined_df = pd.concat(dfs)
combined_df
```

```python
combined_df.to_excel("result.xlsx")
```

```python
import xlwings as xw
import pandas as pd

app = xw.App(visible=True)

workbook = xw.Book("data.xlsx")

dfs = []
for sheet in workbook.sheets:
    df = sheet.range("A2").options(pd.DataFrame,expand="table",index=False).value

    df["해당연도"] = sheet.name

    dfs.append(df)

combined_df_2 = pd.concat(dfs)
```
