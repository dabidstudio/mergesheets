```python
import pandas as pd
import xlwings as xw

## 1단계: 판다스만을 사용하여 Excel 파일 처리

# Excel 파일 열기
excel_file = pd.ExcelFile('data.xlsx')

# DataFrame을 저장할 빈 리스트 초기화
dfs = []

# Excel 파일의 각 시트를 순회
for sheet_name in excel_file.sheet_names:
    # 시트를 DataFrame으로 읽어오기, 헤더가 첫 번째 행에 있다고 가정
    # 헤더가 첫 번째 행이 아닌 경우 이 부분을 조정할 필요가 있음
    df = pd.read_excel(excel_file, header=1, sheet_name=sheet_name)

    # 인덱스를 리셋하고 새로운 열로 시트 이름 추가
    df.reset_index(drop=True, inplace=True)
    df['target_year'] = sheet_name

    # 리스트에 DataFrame 추가
    dfs.append(df)

# 모든 DataFrame을 하나의 DataFrame으로 결합
combined_df2 = pd.concat(dfs, ignore_index=True)

combined_df2
```

```python

## 2단계: xlwings를 사용하여 Excel 파일 처리

# 워크북 열기
workbook = xw.Book('data.xlsx')

# 기존 코드
dfs = []
for sheet in workbook.sheets:
    # 시트에서 DataFrame으로 데이터 읽기, 'A2' 셀부터 'table' 범위까지
    df = sheet.range('A2').options(pd.DataFrame, expand='table', index=False).value
    # 인덱스 리셋하고 시트 이름을 새로운 열로 추가
    df.reset_index(drop=True, inplace=True)
    df['target_year'] = sheet.name
    # 리스트에 DataFrame 추가
    dfs.append(df)

# 모든 DataFrame을 하나의 DataFrame으로 결합
combined_df = pd.concat(dfs, ignore_index=True)
```
